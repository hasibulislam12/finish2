# Input a image and show using python

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
image1 = Image.open('Images/nature.png')
image1_array = np.array(image1)

gray_image2 = image1.convert('L')
image2_array = np.array(gray_image2)

# Display
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(image1_array, cmap=None)
plt.title('Orginal Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(image2_array, cmap='Purples')
plt.title('Gray Image')
plt.axis('off')
```

# Add Square border

```python
height = image1.height
width = image1.width

bordered_image = np.array(image1)
border = 50

for i in range(height):
    for j in range(width):
        if(i<border or i>(height-border) or j<border or j>(width-border)):
            bordered_image[i][j] = 10
                    
plt.imshow(bordered_image, cmap=None)
plt.title('Border Image')
plt.axis('off')
```

# Change the diagonal value

```python
resize_image = gray_image2.resize((500,500))
diagonal_image = np.array(resize_image)

Sum = 0

for i in range(500):
    for j in range(500):
        if i == j:
            Sum += diagonal_image[i, j]
            diagonal_image[i, j] = 0
        if (i + j) == 500-1:
            diagonal_image[i, j] = 0
            
diagonal_image[249, 249] = Sum

print(Sum)
plt.imshow(diagonal_image, cmap='gray')
plt.title('Diagoanl value change')
plt.axis('off')
```
